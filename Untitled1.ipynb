```python
A = {
    0: [1, 2, 3],
    1: [0, 2, 4, 5],
    2: [0, 1, 6],
    3: [0],
    4: [1],
    5: [1],
    6: [2]
}
from collections import deque

def bfs(A):
    # TODO : BFS로 탐색하는 로직을 구현해주세요.
    # 방문한 노드를 순서대로 answer 리스트에 넣어서 리턴해주세요.

    answer = []
    visited = [0]
    queue = deque([0])

    while queue:
      current = queue.pop()
      answer.append(current)

      for w in (A[current]):
        if w not in visited:
          visited.append(w)
          queue.appendleft(w)

    return answer


def dfs(A):
    # TODO : DFS로 탐색하는 로직을 구현해주세요.
    # 방문한 노드를 순서대로 answer 리스트에 넣어서 리턴해주세요.

    answer = [] #방문한 노드 저장
    visited = [0]  #자료 구조에 저장된 노드
    stack = deque([0]) #0번 노드에서 시작

    while stack: # 스택이 비었다면 false 하나라도 있다면 true
        current = stack.pop() #스택에서 제일 마지막 노드를 빼서 저장
        answer.append(current) #꺼낸 노드를 탐색순서 리스트에 저장

        for w in sorted(A[current], reverse = True): #0의 인접한 노드의 번호를 역순으로 w에 대입하여 반복문
          if w not in visited: #노드가 한번도 자료구조에 저장된 적이 없다면?
            visited.append(w) #자료구조에 저장하고 방문할 목록에 추가/아니라면 버리고 스택에서 팝
            stack.append(w) #스택에 추가
    return answer

# 아래는 체크함수입니다. 수정하실 필요 없습니다.
bfs_result = bfs(A)
dfs_result = dfs(A)

assert bfs_result == [0,1,2,3,4,5,6]
assert dfs_result == [0,1,4,5,2,6,3]
print('PASSED!')

```
